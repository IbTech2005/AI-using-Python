1.Write a pyhton program to find the intersection of two list without using built in set functions.

Example

list1 = [1,2,3,4,5]

list2 = [3,4,5,6,7]

output = [3,4,5]

```python
#1
list1 = [1,2,3,4,5]
list2 = [3,4,5,6,7]

result = []

for i in list1:
  if i in list2 and i not in result:
    result.append(i)
print(result)
```

2.write a program to find of pair of elements in a list whose sum equals a given target value.

Example

Input:[1,2,3,4,5,6]

Output:[4,5,6,1,2,3]


```python
#2
lst = [1,2,3,4,5,6]

target = 9

for i in range(len(lst)):
  for j in range(i+1,len(lst)):
    if lst[i] + lst[j] == target:
      print((lst[i] , lst[j]))
```

3.Write a program to split to a list into two halves and swap them.

Example

input:[1,2,3,4,5,6]

Output:[4,5,6,1,2,3]

```python
#3
lst = [1,2,3,4,5,6]

mid = len(lst)//2
result = lst[mid:] + lst[:mid]
print(result)
```

```python
#6
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

lst = [2, 3, 4, 5, 6, 7, 8]
count = sum(1 for i in lst if is_prime(i))
print(count)
```
