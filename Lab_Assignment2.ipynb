1.Given a list of numbers, separate even and odd numbers into two different lists.
ate even and odd numbers into two different lists.
Example
Input: [10, 21, 34, 41, 56]
Output:
Even = [10, 34, 56]
Odd = [21, 41]

```python
#1
list = [10,21,34,41,56]
even = []
odd = []
print(f"The give list is {list}")
for num in list:
  if num % 2 == 0:
    even.append(num)
  else:
    odd.append(num)

print(f"The even numbers are {even}")
print(f"The odd numbers are {odd}")
```

11.Write a program to count the occurrence of each character in a string using dictionary.

12.Write a function isPrime(n) that checks whether a number is prime or not.

```python
#12
def isPrime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

num = int(input("Enter a number: "))

if isPrime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is not a Prime Number")
```

13.Write a function that returns the factorial of a number using recursion.

```python
#13
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

num = int(input("Enter a number: "))

result = factorial(num)

print("Factorial of", num, "is", result)
```

14.Write a function count_vowels(string) that counts the number of vowels in a string.
Example
Input: "Artificial Intelligence"

```python
#14
def count_vowels(string):
    count = 0
    vowels = "aeiouAEIOU"

    for char in string:
        if char in vowels:
            count += 1

    return count


text = input("Enter a string: ")
result = count_vowels(text)

print("Number of vowels:", result)
```

15.Write a function that accepts a list and returns the sum of all elements divisible by 3.

```python
#15
def sum_divisible_by_3(lst):
    total = 0

    for num in lst:
        if num % 3 == 0:
            total += num

    return total

numbers = list(map(int, input("Enter elements separated by space: ").split()))

print(f"The given list is {numbers}")

result = sum_divisible_by_3(numbers)

print("Sum of numbers divisible by 3:", result)
```

16.Write a function that takes a list of numbers and returns a new list containing only prime numbers.
Example
Input: [10,11,12,13,14,17]
Output: [11,13,17]

```python
#16
def isPrime(n):
    if n <= 1:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True


def get_primes(lst):
    prime_list = []

    for num in lst:
        if isPrime(num):
            prime_list.append(num)

    return prime_list


numbers = list(map(int, input("Enter numbers separated by space: ").split()))

result = get_primes(numbers)

print("Prime numbers:", result)
```
