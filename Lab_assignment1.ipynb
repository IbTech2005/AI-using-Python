15. Write a Python program to print the Fibonacci series up to 10 terms.

```python
#15
def feb(n):
    if(n <= 1):
      return n
    else:
      return feb(n-1) + feb(n-2)
n = int(input("Enter the Fibonacci terms : "))
for i in range(n):
    print(f"{feb(i)}")


```

17. Write a Python program to check whether a number is palindrome or not.

```python
#17
num = int(input("Enter the number : "))
org = num
rev = 0
while num > 0:
          digi = num % 10
          rev = rev * 10 + digi
          num = num // 10

if org == rev:
    print("The number is palindrome")
else:
    print("The number is not palindrome")
```

21. Write a Python program to find GCD & LCM of two numbers.

```python
#21
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

x = a
y = b

while y != 0:
    x, y = y, x % y

gcd = x

lcm = (a*b)//gcd

print("GCD =", gcd)
print("LCM =", lcm)
```

22. Write a Python program to check whether a number is Armstrong or not.

```python
#22
num = int(input("Enter number: "))

temp = num
sum = 0

while num > 0:
    digit = num % 10
    sum += digit**3
    num //= 10

if sum == temp:
    print("Armstrong number")
else:
    print("Not Armstrong")
```
